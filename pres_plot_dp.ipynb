```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Set the style
plt.style.use('seaborn-v0_8-white')
sns.set_context("talk")

# Generate data
x = np.linspace(-10, 10, 1000)
y1 = np.exp(-(x - 2)**2 / 4) 
y2 = np.exp(-(x + 2)**2 / 6)

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y1, color='#179c7d', linewidth=3)
sns.lineplot(x=x, y=y2, color='#005b7f', linewidth=3)

# Remove spines and labels
sns.despine(left=True, bottom=True)
# sns.despine()
plt.xticks([])
plt.yticks([])

# Remove grid
plt.grid(True, color='black', alpha=0.1)

plt.tight_layout()
plt.show()
```
