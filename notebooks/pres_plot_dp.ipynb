```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set the style
plt.style.use("seaborn-v0_8-white")
sns.set_context("talk")

# Generate data
x = np.linspace(-10, 10, 1000)
y1 = np.exp(-((x - 2) ** 2) / 4)
y2 = np.exp(-((x + 2) ** 2) / 6)

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y1, color="#179c7d", linewidth=3)
sns.lineplot(x=x, y=y2, color="#005b7f", linewidth=3)

# Remove spines and labels
sns.despine(left=True, bottom=True)
# sns.despine()
plt.xticks([])
plt.yticks([])

# Remove grid
plt.grid(True, color="black", alpha=0.1)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set the style for dark mode
plt.style.use("dark_background")
sns.set_context("talk")

# Generate data
x = np.linspace(-10, 10, 1000)
y1 = np.exp(-((x - 2) ** 2) / 4)
y2 = np.exp(-((x + 2) ** 2) / 6)

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(
    x=x, y=y1, color="#2ecc71", linewidth=3, label=r"$\mathcal{D}$"
)  # Brighter green for dark background
sns.lineplot(
    x=x, y=y2, color="#3498db", linewidth=3, label=r"$\mathcal{D}\prime$"
)  # Brighter blue for dark background

# Remove spines and labels
sns.despine()
# sns.despine(left=True, bottom=True)
# plt.xticks([])
# plt.yticks([])

# Add subtle grid for dark mode
plt.grid(True, color="white", alpha=0.25)

plt.tight_layout()
plt.show()
```
